BleachBoy is a OpenCV based workflow that takes segmented annotated images from CoralVision, an application developed by SJSU professor Philip Heller. CoralVision is a Deep Learning Computer Vision model that uses Mask R-CNN with the purpose of being able to identify corals, and coral supported species within images. Without Professor Heller, the base of this workflow would not only be posssible, but his Good Vibrations, mentorship and guidance was immeasurable.

The purpose of the BleachBoy workflow is to identify healthy from bleached coral individuals based on the mean RGB difference between the bleached/healthy coral and the mean RGB of the global image.


Development was made possible by the dataset scrupulosly collected and provided by Mauricio Lopez Padierna at Qatar University. #Breakdown of Workflow


   
    ┌─────────────────────────────────────────────┐
    │  1. INPUTS                                  │
    │  ├─ Original coral image (RGB/BGR)          │
    │  └─ Mask R-CNN visualization (colored mask) │
    └─────────────────────────────────────────────┘    
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  2. COLOR SPACE CONVERSION (RGB → HSV)      │
    │  Convert visualization to HSV color space   │
    │  → isolates bright, saturated outline pixels│
    │  (Hue, Saturation, Value)                   │
    └─────────────────────────────────────────────┘
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  3. COLOR THRESHOLDING                      │
    │  Apply `cv2.inRange()` to select high       │
    │  saturation/value pixels (mask “ink”)       │
    │  → retains only coral outline colors        │
    └─────────────────────────────────────────────┘
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  4. EDGE CLEANUP & CONTOUR EXTRACTION       │
    │  ├─ Smooth binary mask (median blur)        │
    │  ├─ Detect edges with `cv2.Canny()`         │
    │  ├─ Close gaps using dilation               │
    │  └─ Extract outlines via `cv2.findContours()`│
    └─────────────────────────────────────────────┘
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  5. REGION FILLING                          │
    │  Fill contours with `cv2.drawContours()`    │
    │  (thickness = cv2.FILLED)                   │
    │  → creates solid coral instance masks       │
    └─────────────────────────────────────────────┘
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  6. MEAN COLOR CALCULATION                  │
    │  Use `cv2.mean(image, mask=mask)`           │
    │  to compute mean R, G, B intensity inside   │
    │  each coral region.                         │
    └─────────────────────────────────────────────┘
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  7. BLEACHING CLASSIFICATION                │
    │  Compare coral brightness to global mean:   │
    │   If meanRGB ≥ 1.05 × globalMean → BLEACHED │
    │   Else → UNBLEACHED                         │
    └─────────────────────────────────────────────┘
                        │
                        ▼
    ┌─────────────────────────────────────────────┐
    │  8. OUTPUTS                                 │
    │  ├─ CSV file: mean RGB + bleaching label    │
    │  ├─ Overlay image: labeled coral contours   │
    │  └─ Summary stats (precision, recall, etc.) │
    └─────────────────────────────────────────────┘

#Final Refactoring


```python
# === BleachBoy ===
import cv2, numpy as np, pandas as pd
from pathlib import Path

# ---------------------------------------
# Palette matching
DELTAE_MAX = 34.0       # allow some flexibility in color palette (ΔE in Lab)
SV_Q     = 0.80         # only keep top-20% saturated/bright pixels
SV_FLOOR = 90           # but never go below this

# Local contrast vs 11x11 mean (edge-ness)
LOCAL_WIN    = 11        # 11x11 window for local-contrast check
LOCAL_DE_MIN = 10.0      # raise to cut false positives

# Morphology to make ring continuous (close tiny gaps in pixels)
CLOSE_ITERS      = 2
DILATE_ITERS     = 1
GAP_CLOSE_KERNEL = 3      # extra 3x3 (elliptical) close for final touch

# Hole selection & shape sanity
HOLE_MIN_AREA = 150
HOLE_CIRC_MIN = 0.20
EDGE_MARGIN   = 8         # reserved for edge rejection experiments (not used below)
MIN_CORE_DIST = 6         # require thick interior away from paint
CORE_MIN_AREA = 220       # area of that thick interior

#Edge-closure & coverage rules
# Treat the outer frame as a "ring" to close edge-cut rings, but still require some true-ring contact so frame-only objects don't pass.
COVERAGE_DILATE_PX      = 3
COVERAGE_MIN            = 0.70   # ≥70% of boundary touches (ring OR frame)
MIN_RING_TOUCH_PX       = 100    # absolute count of boundary pixels touching (ring OR frame)
MIN_TRUE_RING_TOUCH_PX  = 40     # must touch TRUE ring (not just frame)

# Keep averages away from paint (when measuring interior)
RING_CLEAR_PX         = 3        # exclude pixels within 3 px of ring paint
INTERIOR_SHRINK_ITERS = 3        # erode interior before averaging (stay off paint)

# Classification (relative to image mean, or absolute thresholds if you switch)
CLASS_MODE     = "relative"  # "relative" or "absolute"
REL_GRAY_DELTA = 12.0        # brighter-than-image-mean by this many gray levels => bleached
REL_MIN_SAT    = 0.0         # optional saturation limit (0 => disabled)
REL_MAX_CHROMA = 255.0       # optional chroma ceiling (255 => effectively disabled)


EXTS = {".png",".jpg",".jpeg",".tif",".tiff",".bmp"}  # supported image types

# Neon palette of the confidence outlines in CoralVision (BGR!)
PALETTE_BGR = np.array([
  [  0,   0, 255],   # red
  [  0, 128, 255],   # orange
  [  0, 165, 255],   # deep orange
  [  0, 200, 255],   # strong yellow-orange
  [  0, 255, 255],   # yellow
  [  0, 255,   0],   # green
  [255, 255,   0],   # cyan
  [255, 128,   0],   # light sky blue
  [255,   0,   0],   # blue
  [190,   0,   0],   # dark blue
  [140,   0,   0],   # darker blue
], dtype=np.uint8)

# Error and file I/o
def _read_bgr(p: Path):
  im = cv2.imread(str(p), cv2.IMREAD_COLOR)
  if im is None:
    raise RuntimeError(f"Cannot read: {p}")
  return im  # image as NumPy array in OpenCV BGR format

def bgr_to_lab(img_bgr):
  return cv2.cvtColor(img_bgr, cv2.COLOR_BGR2LAB).astype(np.float32)

def deltaE_cie76(lab1, lab2):
  diff = lab1 - lab2
  return np.sqrt((diff**2).sum(axis=2))

# ring detector (binary mask: 1 = ring pixel)
#Neon ring detector:
#1) HSV gate (bright & saturated),
#2) Lab ΔE to neon palette,
#3) local Lab contrast vs 11x11 neighborhood
#4) morphology to close gaps.

def robust_ring_mask(bgr):
  h, w = bgr.shape[:2]

  # 1) HSV gate (keep top-20% in S/V but not below floor)
  hsv = cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV).astype(np.float32)
  S, V = hsv[...,1], hsv[...,2]
  s_cut = max(np.quantile(S, SV_Q), SV_FLOOR)
  v_cut = max(np.quantile(V, SV_Q), SV_FLOOR)
  gate = ((S >= s_cut) & (V >= v_cut)).astype(np.uint8)

  # 2) Palette match in Lab (closest ΔE to any palette color)
  lab = bgr_to_lab(bgr)
  de_min = np.full((h, w), 1e9, np.float32)
  for bgr_ref in PALETTE_BGR:
    # FIX: ensure a proper 1x1x3 uint8 "image" for cvtColor
    lab_ref = bgr_to_lab(np.array([[bgr_ref]], dtype=np.uint8))[0, 0, :]
    de = deltaE_cie76(lab, lab_ref)
    de_min = np.minimum(de_min, de)
  pal_mask = (de_min <= DELTAE_MAX).astype(np.uint8)

  # 3) Edges pop vs 11x11 blur
  k = LOCAL_WIN | 1                     # force odd kernel
  lab_blur = cv2.blur(lab, (k, k))
  de_loc = np.sqrt(((lab - lab_blur)**2).sum(axis=2))
  local_ok = (de_loc >= LOCAL_DE_MIN).astype(np.uint8)

  # Combine gates
  m = gate & pal_mask & local_ok

  # 4) Morphology: close tiny gaps, slightly thicken if needed
  if CLOSE_ITERS:
    m = cv2.morphologyEx(m, cv2.MORPH_CLOSE, np.ones((3,3), np.uint8), iterations=CLOSE_ITERS)
  if DILATE_ITERS:
    m = cv2.dilate(m, np.ones((3,3), np.uint8), iterations=DILATE_ITERS)
  if GAP_CLOSE_KERNEL and GAP_CLOSE_KERNEL > 1:
    k2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (GAP_CLOSE_KERNEL, GAP_CLOSE_KERNEL))
    m  = cv2.morphologyEx(m, cv2.MORPH_CLOSE, k2, iterations=1)

  return (m > 0).astype(np.uint8)

# add a 1px frame so edge-cut rings are closed
def add_frame(mask):
  m = mask.copy()
  m[0, :]  = 1; m[-1, :] = 1; m[:, 0] = 1; m[:, -1] = 1
  return m
# shape helper: 1.0 = perfect circle
def circularity(cnt):
  A = cv2.contourArea(cnt)
  if A <= 0:
    return 0.0
  P = max(cv2.arcLength(cnt, True), 1e-6)
  return float(4.0 * np.pi * A / (P * P))

#coverage test: does the hole boundary touch ring/frame enough; still not great.
def ring_coverage(hole_cnt, ring_mask, frame_mask, dilate_px=COVERAGE_DILATE_PX):
  h, w = ring_mask.shape
  # thin 1-px boundary of the hole
  boundary = np.zeros((h, w), dtype=np.uint8)
  cv2.drawContours(boundary, [hole_cnt], -1, 1, thickness=1)
  # dilate ring & frame a little to allow tolerance
  k = np.ones((int(dilate_px), int(dilate_px)), dtype=np.uint8)
  ring_u8 = (ring_mask  > 0).astype(np.uint8)
  frame_u8= (frame_mask > 0).astype(np.uint8)
  ring_d = cv2.dilate(ring_u8,  k, iterations=1)
  frame_d= cv2.dilate(frame_u8, k, iterations=1)
  ring_or_frame = cv2.bitwise_or(ring_d, frame_d)
  touch_any = cv2.bitwise_and(boundary, ring_or_frame)
  touch_ring = cv2.bitwise_and(boundary, ring_d)
  total = int(boundary.sum())
  frac_any = float(touch_any.sum()) / max(total, 1)
  touch_any_n= int(touch_any.sum())
  touch_ring_n= int(touch_ring.sum())
  return (frac_any, touch_any_n, touch_ring_n)

def distance_from_ring(ring_or_frame):
  inv = (ring_or_frame == 0).astype(np.uint8)  # distanceTransform expects 0 as "foreground"
  return cv2.distanceTransform(inv, cv2.DIST_L2, 5)
def hole_regions_from_border(ring_mask):
  H, W = ring_mask.shape
  frame = np.zeros_like(ring_mask, np.uint8)
  frame[0, :]=frame[-1, :]=frame[:, 0]=frame[:, -1]=1
  ring_for_ct= add_frame(ring_mask)                 # for contour topology (closes edge shapes)
  ring_for_dt= (ring_mask | frame).astype(np.uint8) # for distance map (ring+frame)
  # Find contours with parent/child hierarchy
  mask_u8  = (ring_for_ct * 255).astype(np.uint8)
  contours, hier = cv2.findContours(mask_u8, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
  out = []
  if hier is None or len(contours) == 0:
    return out
  hier = hier[0]
  # Distance map used to define 'clear' (far from paint) and 'core' (thick interior)
  ring_dist= distance_from_ring(ring_for_dt)
  # Iterate parents (outer rings), then their children (holes)
  for i, c in enumerate(contours):
    if hier[i][3] != -1:  # skip non-parents
      continue
    child = hier[i][2]
    while child != -1:
      hole = contours[child]
      area = cv2.contourArea(hole)
      if area < HOLE_MIN_AREA or circularity(hole) < HOLE_CIRC_MIN:
        child = hier[child][0]; continue
      # Build hole interior, shrink away from edges
      hole_mask = np.zeros((H, W), np.uint8)
      cv2.drawContours(hole_mask, [hole], -1, 1, thickness=cv2.FILLED)
      if INTERIOR_SHRINK_ITERS:
        hole_mask = cv2.erode(hole_mask, np.ones((3,3), np.uint8), iterations=INTERIOR_SHRINK_ITERS)
      if hole_mask.sum() == 0:
        child = hier[child][0]; continue
      # Coverage vs ring+frame; also require some TRUE ring contact
      frac_any, touch_any, touch_true = ring_coverage(hole, ring_mask, frame, COVERAGE_DILATE_PX)
      if not (frac_any >= COVERAGE_MIN and touch_any >= MIN_RING_TOUCH_PX and touch_true >= MIN_TRUE_RING_TOUCH_PX):
        child = hier[child][0]; continue
      # 'Clear' interior: keep pixels ≥ RING_CLEAR_PX from ring/frame
      clear = (hole_mask & (ring_dist >= RING_CLEAR_PX)).astype(np.uint8)
      if clear.sum() < HOLE_MIN_AREA:
        child = hier[child][0]; continue
      # 'Core' thickness: require enough pixels ≥ MIN_CORE_DIST (kill skinny lens overlaps)
      core = (hole_mask & (ring_dist >= MIN_CORE_DIST)).astype(np.uint8)
      if core.sum() < CORE_MIN_AREA:
        child = hier[child][0]; continue
      out.append((hole, clear))
      child = hier[child][0]
  return out
# per-image baseline stats for relative classification
def image_stats(bgr):
  gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
  return dict(img_mean_gray=float(gray.mean()), img_std_gray =float(gray.std()))
# metrics for one hole interior used by classifirs
def compute_metrics(bgr, interior):
  if interior.sum() == 0:
    return None
  B, G, R = cv2.split(bgr)
  gray= cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
  hsv= cv2.cvtColor(bgr, cv2.COLOR_BGR2HSV)
  S= hsv[..., 1]
  sel= interior.astype(bool)
  mR   = float(R[sel].mean()); mG = float(G[sel].mean()); mB = float(B[sel].mean())
  mGray= float(gray[sel].mean()); mSat = float(S[sel].mean())
  chroma = float(max(mR, mG, mB) - min(mR, mG, mB))
  return dict(mean_R=mR, mean_G=mG, mean_B=mB,mean_gray=mGray, mean_sat=mSat,chroma=chroma, area_px=int(sel.sum()))

#classifier
def classify_relative(m, stats):
  bright = (m["mean_gray"] >= stats["img_mean_gray"] + REL_GRAY_DELTA)
  sat_ok = (m["mean_sat"]  <= REL_MIN_SAT) if REL_MIN_SAT > 0 else True
  chr_ok = (m["chroma"]    <= REL_MAX_CHROMA)
  return "bleached" if (bright and sat_ok and chr_ok) else "unbleached"

def classify_absolute(m):
  ABS_GRAY_THR   = 172.0
  ABS_SAT_MAX    = 125.0
  ABS_CHROMA_MAX = 55.0
  return "bleached" if (m["mean_gray"]>=ABS_GRAY_THR and m["mean_sat"] <=ABS_SAT_MAX and m["chroma"]   <=ABS_CHROMA_MAX) else "unbleached"

# --- overlay renderer (no fill, contour + label) ---
def overlay_labels_no_fill(bgr, hole_regions, labels):
  out = bgr.copy()
  for idx, ((hole_cnt, _), lab) in enumerate(zip(hole_regions, labels), 1):
    cv2.drawContours(out, [hole_cnt], -1, (255,255,255), 2)   # white outline
    x, y, w, h = cv2.boundingRect(hole_cnt)
    tcol = (0,0,255) if lab == "bleached" else (0,255,0) # red=bleached, green=unbleached
    cv2.putText(out, f"{idx}:{lab}", (x, max(25, y-6)), cv2.FONT_HERSHEY_SIMPLEX, 0.8, tcol, 2, cv2.LINE_AA)
  return out

#folder drie
def analyze_image(p: Path, out_dir: Path):
  img   = _read_bgr(p)
  stats = image_stats(img)
  ring  = robust_ring_mask(img)# palete mao
  holes = hole_regions_from_border(ring)# valid hole contours & clear interiors
  rows, labels = [], []
  for i, (hole_cnt, mask) in enumerate(holes, 1):
    m = compute_metrics(img, mask)
    if not m:
      continue
    lab = classify_relative(m, stats) if CLASS_MODE == "relative" else classify_absolute(m)
    labels.append(lab)
    rows.append({"image": p.name, "object_id": i, "classification": lab, **{k: (round(v, 3) if isinstance(v, (int, float)) else v) for k, v in m.items()},**{k: round(v, 3) for k, v in stats.items()}})

  if holes:
    (out_dir / "overlays").mkdir(parents=True, exist_ok=True)
    ov = overlay_labels_no_fill(img, holes, labels)
    cv2.imwrite(str((out_dir / "overlays" / f"{p.stem}_overlay.png")), ov)
  return rows
# Savinf results and foler results
def analyze_folder(INPUT_DIR: Path, REPORT_DIR: Path):
  REPORT_DIR.mkdir(parents=True, exist_ok=True)
  imgs = sorted([p for p in INPUT_DIR.iterdir() if p.suffix.lower() in EXTS])
  all_rows = []
  for p in imgs:
    all_rows += analyze_image(p, REPORT_DIR)
    print(f"{p.name}: processed")
  df = pd.DataFrame(all_rows)
  out_csv = REPORT_DIR / "summary.csv"
  df.to_csv(out_csv, index=False)
  print("Saved:", out_csv)
  if not df.empty:
    counts = df.groupby("image")["classification"].value_counts().unstack(fill_value=0)
    counts["total"] = counts.sum(axis=1)
    print("\nPer-image counts:\n", counts)

  return df
```
